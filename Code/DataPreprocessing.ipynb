```python
#importing required libraries
import pandas as pd
```

```python
# reading the disease-symptoms dataset
disease_df = pd.read_csv(r"C:\Users\patel\OneDrive\Desktop\AI\Project\DataSet\dataset.csv")
disease_df.head()
```

```python
# checking for null values
disease_df.isnull().sum()
```

```python
# replacing the null values with 0
disease_df = disease_df.fillna(0)
disease_df.head()
```

```python
# removing the symptoms containing large number of null values
del disease_df['Symptom_8']
del disease_df['Symptom_9']
del disease_df['Symptom_10']
del disease_df['Symptom_11']
del disease_df['Symptom_12']
del disease_df['Symptom_13']
del disease_df['Symptom_14']
del disease_df['Symptom_15']
del disease_df['Symptom_16']
del disease_df['Symptom_17']
```

```python
# printing the dataset after the deletion of the columns
disease_df.head()
```

```python
# creating a dictionary to store list of symptoms related to that disease 
diseaseDict = disease_df.set_index('Disease').T.to_dict('list')
print(diseaseDict)
```

```python
# storing the dict as dataframe with 'Disease' and 'Symptom' columns
disease_symptom_df = pd.DataFrame(list(diseaseDict.items()), columns=['Disease','Symptom'])
disease_symptom_df.head()
```

```python
# getting all the unique diseases present in the dataset
print("Unique Diseases: ", disease_symptom_df['Disease'].nunique())
disease_symptom_df['Disease'].unique()
```

```python
# storing the dataframe to a csv file 
disease_symptom_df.to_csv('C:/Users/patel/OneDrive/Desktop/AI/Project/DataSet/Disease_Symptoms_Combined.csv',index=False)
```

```python
# reading the drug dataset
drug_df = pd.read_csv(r"C:\Users\patel\OneDrive\Desktop\AI\Project\DataSet\drugs_dataset.csv")
drug_df.head()
```

```python
# droping the unnecessary columns
drug_df.drop(['uniqueID', 'date'], axis = 1, inplace=True)
```

```python
# remaning the condition column to disease column for better understanding
drug_df = drug_df.rename(columns={"condition": "Disease"})
```

```python
# droping the data rows containing null values and then sorting them by disease
drug_df.dropna(axis = 0, inplace=True)
drug_df.sort_values(by = ['Disease'], ascending= False , inplace= True)
drug_df.head()
```

```python
# getting all the unique disease present in the drug dataset  
print("Unique Diseases: ", drug_df['Disease'].nunique())
drug_df['Disease'].unique()
```

```python
# getting all the unique disease present in the drug dataset  
print("Unique Drugs: ", drug_df['drugName'].nunique())
drug_df['drugName'].unique()
```

```python
# droping the data rows containing null values and then sorting them by disease
disease_symptom_df.dropna(axis = 0, inplace=True)
disease_symptom_df.sort_values(by = ['Disease'], ascending= False , inplace= True)
disease_symptom_df.head()
```

```python
# converting all the disease names from uppercase to lowercase
drug_df['Disease'] = drug_df['Disease'].str.lower()
disease_symptom_df['Disease'] = disease_symptom_df['Disease'].str.lower()
```

```python
disease_symptom_df.head()
```

```python
# combining the disease-symptom dataset with the drug dataset based on the similar diseases
final_dataset = pd.merge(drug_df, disease_symptom_df, on='Disease')
```

```python
# printing the merged dataset
final_dataset.head()
```

```python
# getting no. of rows and columns present in the dataset
final_dataset.shape
```

```python
# renaming the columns for better understanding
final_dataset.columns = ['Drug', 'Disease', 'Review','Rating' ,'UsefulCount', 'Symptoms']
```

```python
# getting the no. of unique diseases present in the combined dataset 
print("No. of unique diseases: ", final_dataset['Disease'].nunique())
final_dataset['Disease'].unique()
```

```python
# getting the no. of unique drugs present in the combined dataset
print("No. of unique drugs: ", final_dataset['Drug'].nunique())
final_dataset['Drug'].unique()
```

```python
# printing the final combined dataset
final_dataset.head()
```

```python
# storing the combined dataset to the csv 
final_dataset.to_csv('C:/Users/patel/OneDrive/Desktop/AI/Project/DataSet/DiseaseDrug_dataset.csv',index=False)
```
